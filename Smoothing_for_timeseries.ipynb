```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/Data_used_for_practice/monthly-beer-production-in-austr.csv', parse_dates=True)
df.head()
```

```python
df1 = df.copy()
```

```python
df1.plot()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
%matplotlib inline
```

```python
df1.dropna(inplace= True)
```

Simple moving average


```python
# calculating simple moving average
# using .rolling(window).mean() ,
# with window size = 3
df1['SMA3'] = df1['Monthly beer production'].rolling(3).mean()
```

```python
df1[['Monthly beer production', 'SMA3']].plot(label='MBP', 
                                  figsize=(16, 8))
```

```python
i = 5 ##i is the size of the moving average

while i<=30:
  df1[f'SMA{i}'] = df1['Monthly beer production'].rolling(i).mean()
  df1[['Monthly beer production', f'SMA{i}']].plot(label='MBP', 
                                  figsize=(16, 8))
  i += 5
```

Exponential Moving Average

```python
df2 = df.copy()
```

```python
i = 5 ##i is the size of the moving average

while i<=30:
  df2[f'EMA{i}'] = df2['Monthly beer production'].ewm(span=i).mean()
  df2[['Monthly beer production', f'EMA{i}']].plot(label='MBP', 
                                  figsize=(16, 8))
  i += 5
```

Cumulative moving average

```python
df3 = df.copy()
```

```python
df3['CMA'] = df3['Monthly beer production'].expanding().mean()
df3[['Monthly beer production', 'CMA']].plot(label='MBP', 
                                  figsize=(16, 8))
```
